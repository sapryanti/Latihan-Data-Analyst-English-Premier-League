```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matolotlib inline
```

```python
loaddataset = pd.read_csv('EPL_20_21.csv')
loaddataset
```

```python
loaddataset.info()
```

```python
loaddataset.describe()
```

```python
loaddataset.isna().sum()
```

```python
#create 2 column
loaddataset['MinsPerMatch'] = (loaddataset['Mins'] / loaddataset['Matches']).astype(int)
loaddataset['GoalsPerMatch'] = (loaddataset['Goals'] / loaddataset['Matches']).astype(float)
loaddataset.head()
```

```python
#menghitung total goals

Total_goals = loaddataset['Goals'].sum()
print(Total_goals)
```

```python
#penalty Goals

Total_penalty_Goals = loaddataset['Penalty_Goals'].sum()
print(Total_penalty_Goals)
```

```python
#penalty Attempts

Total_PenaltyAttempts = loaddataset['Penalty_Attempted'].sum()
print(Total_PenaltyAttempts )
```

```python
#visualisasi chart for penalty missed vs scored

plt.figure(figsize=(13,6))
pl_not_scored = loaddataset['Penalty_Attempted'].sum() - Total_penalty_Goals
data = [pl_not_scored,Total_penalty_Goals]
labels = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('rocket')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
#most players from with countries

nationality =loaddataset.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar',figsize=(12,6), color = sns.color_palette("magma"))
```

```python
#club with maximum players in their squad

loaddataset['Club'].value_counts().nlargest(5).plot(kind = 'bar', color=sns.color_palette('viridis'))
```

```python
#clubs with least players in their squad
loaddataset['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color=sns.color_palette('pastel'))
```

```python
#players based on age group

under20 = loaddataset[loaddataset['Age']<= 20]
age20_25 = loaddataset[(loaddataset['Age'] > 20) & (loaddataset['Age'] <= 25)]
age25_30 = loaddataset[(loaddataset['Age'] > 25) & (loaddataset['Age'] <= 30)]
above30 = loaddataset[loaddataset['Age']>30]
```

```python
x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
mylabels = ["<=20", ">20",">25 7 <= 30", ">30"]
plt.title('Total players with age group', fontsize = 20)
plt.pie(x,labels = mylabels, autopct = "%.lf%%")
plt.show()
```

```python
#total under 20 players in each club

players_under_20 = loaddataset[loaddataset['Age']<20]
players_under_20['Club'].value_counts().plot(kind='bar',color = sns.color_palette("cubehelix"))
```
